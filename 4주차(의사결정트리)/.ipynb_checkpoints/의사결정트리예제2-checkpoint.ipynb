```python
import pandas as pd

train = pd.read_csv("data/cearmic_train.csv", index_col=["id"])
print(train.shape)
train.head()
```

```python
test = pd.read_csv("data/cearmic_test.csv", index_col=["id"])
print(test.shape)
test.head()
```

<h3>데이터 로드 끝</h3>
<h3>전처리 시작</h3>

```python
#측정 불가 행은 삭제시킨다.
train = train[train.average_thickness!="측정불가"]
test = test[test.average_thickness!="측정불가"]
```

```python
#전처리 과정 (인코딩), 0은 이상없다, 1은 이상있다.
train.loc[pd.isnull(train["failure"]), "failure"] = 0
train.loc[train["failure"]!=0,"failure"]=1
```

```python
train
```

```python
test.shape
```

```python
#전처리 과정 (인코딩)
test.loc[pd.isnull(test["failure"]), "failure"] = 0
test.loc[test["failure"]!=0,"failure"]=1
```

<h3>전처리 끝</h3>

<h3>의사결정 트리 생성</h3>

```python
feature_names = ["Gap", "average_thickness", "standard_deviation",
                 "lab_temperature", "lab_humidity"]
```

```python
X_train = train[feature_names]

print(X_train.shape)
X_train.head()
```

```python
#정답셋 준비
label_name = "failure"

y_train = train[label_name]

print(y_train.shape)
y_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
seed = 37
model = DecisionTreeClassifier(max_depth=5,
                               random_state=seed)
model.fit(X_train, y_train)
```

```python
from sklearn.tree import export_graphviz
import graphviz

export_graphviz(model,
                feature_names=feature_names,
                class_names=["No_Failure", "Yes_Failure"],
                out_file="ceramic_decision-tree.dot")

with open("ceramic_decision-tree.dot") as f:
    dot_graph = f.read()
    
graphviz.Source(dot_graph)
```

<h3>의사결정 트리 생성 끝</h3>

<h3>예측 시작</h3>

```python
#예측
X_test = test[feature_names]
prediction = model.predict(X_test)
```

```python
prediction
```

```python
real_result = test["failure"]
real_result
```

```python
def calculate_accuracy():
    amount = len(real_result)-1
    collect_cnt=0
    wrong_cnt=0
    
    for i in range (0,amount):
        if real_result.iloc[i]==prediction[i]:
            collect_cnt+=1
        else:
            wrong_cnt+=1
    
    accuracy = collect_cnt/amount *100
    
    print("테스트 개수: ",amount)
    print("맞은 개수: ",collect_cnt)
    print("틀린 개수: ",wrong_cnt)
    return accuracy
```

```python
calculate_accuracy()
```
