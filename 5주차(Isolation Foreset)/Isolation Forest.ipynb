```python
import pandas as pd 
from imblearn.under_sampling import RandomUnderSampler
from collections import Counter


credit_data = pd.read_csv('mongodb.csv')
credit_data = credit_data[credit_data["평균 두께"]!="측정불가"]
credit_data.loc[pd.isnull(credit_data["비고"]), "비고"] = 0
credit_data.loc[credit_data["비고"]!=0, "비고"] = 1 #이상있는건 1로
credit_data['평균 두께'].astype('float')
credit_data['표준 편차'].astype('float')
credit_data['Gap'].astype('float')
credit_data['두께1'].astype('float')
credit_data['두께2'].astype('float')
credit_data['두께3'].astype('float')
credit_data['두께4'].astype('float')
credit_data['두께5'].astype('float')
```

```python
display(credit_data)
```

```python
#'두께1','두께2','두께3','두께4','두께5'
learning_data = credit_data[['P/B Ratio','바인더 함량','소포제 함량','점도','코팅 속도','실내 온도','실내 습도','Gap','두께1',
                             '두께2','두께3','두께4','두께5','평균 두께','표준 편차','시트 두께','표준 편차.1','비고']]
learning_data = learning_data.reset_index(drop=True)
display(learning_data)
```

```python
from sklearn.ensemble import IsolationForest

clf = IsolationForest(n_estimators=300, contamination=0.60, random_state=1)
```

```python
clf.fit(learning_data)
```

```python
y_pred_outliers = clf.predict(learning_data)
```

```python
out = pd.DataFrame(y_pred_outliers)
```

```python
out = out.rename(columns={0: "out"})
out = pd.DataFrame(out["out"]).replace({-1:0,1:1})
```

```python
race_an1 = pd.concat([learning_data, out], 1)
```

```python
display(race_an1)
```

```python
tt=0
tf=0
ft=0
ff=0
for i in race_an1.index:
    if race_an1.loc[i,'out']==0 and race_an1.loc[i,'비고']==0:
        tt+=1
    elif race_an1.loc[i,'out']==1 and race_an1.loc[i,'비고']==0:
        tf+=1
    elif race_an1.loc[i,'out']==0 and race_an1.loc[i,'비고']==1:
        ft+=1
    elif race_an1.loc[i,'out']==1 and race_an1.loc[i,'비고']==1:
        ff+=1

print("정상인걸 정상이라고 판단 개수:",tt)
print("정상인걸 이상이라고 판단 개수:",tf)
print("이상인걸 정상이라고 판단 개수:",ft)
print("이상인걸 이상이라고 판단 개수:",ff)
```
